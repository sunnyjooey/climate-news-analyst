# This notebook looks at the results of the SVO extraction resulting from the pipeline

```python
import pandas
svo_df = pandas.read_pickle('svo_simplified.pkl')
```

There is always more cleaning to be done.
This particular one ensures there are no weird characters.

```python
import string
exclude = set(string.punctuation + '¦—–…•’‘”“')

svo_df.fillna('', inplace=True)

svo_df['subject'] = svo_df['subject'].apply(lambda x: ''.join(ch for ch in x if ch not in exclude))
svo_df['object'] = svo_df['object'].apply(lambda x: ''.join(ch for ch in x if ch not in exclude))
svo_df['verb'] = svo_df['verb'].apply(lambda x: ''.join(ch for ch in x if ch not in exclude))
svo_df['subject'] = svo_df['subject'].apply(lambda x: x.strip().lower())
svo_df['verb'] = svo_df['verb'].apply(lambda x: x.strip().lower())
svo_df['object'] = svo_df['object'].apply(lambda x: x.strip().lower())

svo_df_noblank = svo_df
for col in ['subject', 'verb', 'object']:
    svo_df_noblank = svo_df_noblank[svo_df_noblank[col] != '']

svo_df_nopn = svo_df_noblank[svo_df_noblank['subject'].apply(
    lambda x: x not in ['be', 's', 'we', 'he', 'she', 'it', 'they'])]
```

```python
pandas.options.display.max_rows = 999
df = svo_df_nopn.groupby(by=['subject', 'verb', 'object'])['country'].value_counts().rename('counts').sort_values(ascending=False).reset_index()
```
